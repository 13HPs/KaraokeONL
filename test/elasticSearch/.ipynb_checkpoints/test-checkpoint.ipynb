```python
def create_index(index_name, settings_mappings):
    if es.indices.exists(index=index_name):
        es.indices.delete(index=index_name)
    es.indices.create(index=index_name, body=settings_mappings)
```

```python
lyrics_settings = {
    "settings": {
        "analysis": {
            "analyzer": {
                "vietnamese_analyzer": {
                    "type": "custom",
                    "tokenizer": "icu_tokenizer",
                    "filter": [
                        "lowercase",
                        "icu_folding",
                        "vietnamese_stop",
                        "ngram_filter"
                    ]
                }
            },
            "filter": {
                "vietnamese_stop": {
                    "type": "stop",
                    "stopwords": "_vietnamese_"
                },
                "ngram_filter": {
                    "type": "edge_ngram",
                    "min_gram": 2,
                    "max_gram": 20
                }
            }
        }
    },
    "mappings": {
        "properties": {
            "lyric": {
                "type": "text",
                "analyzer": "vietnamese_analyzer",
                "search_analyzer": "vietnamese_analyzer",
                "fields": {
                    "keyword": {
                        "type": "keyword"
                    }
                }
            },
            "music_id": {
                "type": "keyword"
            }
        }
    }
}

create_index("lyrics", lyrics_settings)
```

```python
def bulk_index_lyrics(lyrics_data, batch_size=100):
    actions = [
        {
            "_index": "lyrics",
            "_id": f"{lyric['music_id']}_{i}",
            "_source": lyric
        }
        for i, lyric in enumerate(lyrics_data)
    ]
    helpers.bulk(es, actions, chunk_size=batch_size)

bulk_index_lyrics(lyrics)
```

```python
# Bulk index accounts data
def bulk_index_accounts(accounts_data):
    actions = [
        {
            "_index": "accounts",
            "_id": account['id'],
            "_source": account
        }
        for account in accounts_data
    ]
    helpers.bulk(es, actions)

# Index the accounts
bulk_index_accounts(accounts)
```

```python
def bulk_index_musics(musics_data, batch_size=100):
    actions = [
        {
            "_index": "musics",
            "_id": music['id'],
            "_source": music
        }
        for music in musics_data
    ]
    helpers.bulk(es, actions, chunk_size=batch_size)

bulk_index_musics(musics)
```

```python
def search_lyrics(query):
    body = {
        "query": {
            "match": {
                "lyric": {
                    "query": query,
                    "fuzziness": "AUTO"
                }
            }
        },
        "collapse": {
            "field": "lyric.keyword"
        },
        "size": 5
    }
    res = es.search(index="lyrics", body=body)
    return [(hit["_source"]["lyric"], hit["_score"]) for hit in res["hits"]["hits"]]
```

```python
def search_musics(query):
    body = {
        "query": {
            "match": {
                "name": {
                    "query": query,
                    "fuzziness": "AUTO"
                }
            }
        },
        "size": 5
    }
    res = es.search(index="musics", body=body)
    
    return [(hit["_source"]["name"], hit["_score"], hit["_source"]["id"]) for hit in res["hits"]["hits"]]

def search_accounts(query):
    body = {
        "query": {
            "match": {
                "name": {
                    "query": query,
                    "fuzziness": "AUTO"
                }
            }
        },
        "size": 5
    }
    res = es.search(index="accounts", body=body)
    return [(hit["_source"]["name"],hit["_source"]["id"] ,hit["_score"]) for hit in res["hits"]["hits"]]
```
